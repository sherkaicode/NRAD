### Boosted Decision Tree

```python
print('Size of data: {}'.format(data.shape))
print('Number of events: {}'.format(data.shape[0]))
print('Number of columns: {}'.format(data.shape[1]))

print ('\nList of features in dataset:')
for col in data.columns:
    print(col)
```

```python
# look at column labels --- notice last one is "Label" and first is "EventId" also "Weight"
print('Number of signal events: {}'.format(len(data[data.Labels == 's'])))
print('Number of background events: {}'.format(len(data[data.Labels == 'b'])))
print('Fraction signal: {}'.format(len(data[data.Labels == 's'])/(float)(len(data[data.Labels == 's']) + len(data[data.Labels == 'b']))))
```

```python
n_train = int(0.8*(len(data[data.Labels == 's']) + len(data[data.Labels == 'b'])))
# n_train = int(0.8*9009000)
print(n_train)

data_train = data[:n_train]
data_test = data[n_train:]
```

```python
print('Number of training samples: {}'.format(len(data_train)))
print('Number of testing samples: {}'.format(len(data_test)))

print('\nNumber of signal events in training set: {}'.format(len(data_train[data_train.Labels == 's'])))
print('Number of background events in training set: {}'.format(len(data_train[data_train.Labels== 'b'])))
print('Fraction signal: {}'.format(len(data_train[data_train.Labels == 's'])/(float)(len(data_train[data_train.Labels == 's']) + len(data_train[data_train.Labels == 'b']))))
```

```python
feature_names = data.columns[:-2].tolist()
feature_names = [str(f) for f in feature_names]# we skip the first and last two columns because they are the ID, weight, and label
train = xgb.DMatrix(data=data_train[feature_names],label=data_train.Labels.cat.codes,missing=-999.0,feature_names=feature_names)
test = xgb.DMatrix(data=data_test[feature_names],label=data_test.Labels.cat.codes, missing=-999.0,feature_names=feature_names)
```

```python
print('Number of training samples: {}'.format(train.num_row()))
print('Number of testing samples: {}'.format(test.num_row()))

print('\nNumber of signal events in training set: {}'.format(len(np.where(train.get_label())[0])))
```

```python
param = {}

# # Booster parameters
param['eta']              = 0.1 # learning rate
param['max_depth']        = 6  # maximum depth of a tree
param['subsample']        = 0.8 # fraction of events to train tree on
param['colsample_bytree'] = 0.8 # fraction of features to train tree on
# param['max_delta_step']   = 1     # Helps stabilize updates for highly imbalanced data

# param['min_child_weight'] = 50
# param['scale_pos_weight'] = 1000 
# Learning task parameters
param['objective']   = 'binary:logistic' # objective function
param['eval_metric'] = 'error'           # evaluation metric for cross validation
param = list(param.items()) + [('eval_metric', 'logloss')] + [('eval_metric', 'rmse')]

# Trying different Parameters
# param['eta']              = 0.05  # Lower learning rate for better convergence
# param['max_depth']        = 6     # Reduce depth to prevent overfitting
# param['subsample']        = 0.7   # Reduce to introduce randomness and prevent overfitting
# param['colsample_bytree'] = 0.7   # Reduce to introduce feature randomness
# param['scale_pos_weight'] = 1000  # Adjust for extreme imbalance
# param['max_delta_step']   = 1     # Helps stabilize updates for highly imbalanced data
# param['min_child_weight'] = 50
# param['objective']   = 'binary:logistic'  # Standard objective for classification
# param['eval_metric'] = 'aucpr'            # PR AUC is better for imbalanced datasets
# param = list(param.items()) + [('eval_metric', 'logloss')] + [('eval_metric', 'rmse')]


num_trees = 1000  # number of trees to make
```

```python
booster = xgb.train(param,train,num_boost_round=num_trees)
```

```python
print(booster.eval(test))
```

```python
predictions = booster.predict(test)
```

```python
# plot all predictions (both signal and background)
plt.figure();
plt.hist(predictions,bins=np.linspace(0,1,50),histtype='step',color='darkgreen',label='All events');
# make the plot readable
plt.xlabel('Prediction from BDT',fontsize=12);
plt.ylabel('Events',fontsize=12);
plt.legend(frameon=False);

# plot signal and background separately
plt.figure();
plt.hist(predictions[test.get_label().astype(bool)],bins=np.linspace(0,1,50),
         histtype='step',color='midnightblue',label='signal');
plt.hist(predictions[~(test.get_label().astype(bool))],bins=np.linspace(0,1,50),
         histtype='step',color='firebrick',label='background');
# make the plot readable
plt.yscale('log')
plt.xlabel('Prediction from BDT',fontsize=12);
plt.ylabel('Events',fontsize=12);
plt.legend(frameon=False);
```

```python
# choose score cuts:
cuts = np.linspace(0,1,500);
nsignal = np.zeros(len(cuts));
nbackground = np.zeros(len(cuts));
for i,cut in enumerate(cuts):
    nsignal[i] = len(np.where(predictions[test.get_label().astype(bool)] > cut)[0]);
    nbackground[i] = len(np.where(predictions[~(test.get_label().astype(bool))] > cut)[0]);
    
# plot efficiency vs. purity (ROC curve)
plt.figure();
plt.plot(nsignal/len(data_test[data_test.Labels == 's']),nsignal/(nsignal + nbackground),'o-',color='blueviolet');
# # make the plot readable
# # X-Axis is Recall
# # Y-Axis is TP/(TP + TN)
plt.xlabel('Efficiency',fontsize=12);
plt.ylabel('Purity',fontsize=12);
plt.legend(frameon=False);


num = np.divide(nsignal, len(data_test[data_test.Labels == 's']))
deno = np.divide(nsignal,(nsignal + nbackground), where=(nsignal+nbackground)!=0)
#deno = np.divide(nsignal,(nsignal + nbackground))
#plt.plot(num,deno,color='blue', alpha=.8);
# plt.plot(nsignal/len(data_test[data_test.Labels == 's']),nsignal/(nsignal + nbackground),color='midnightblue', alpha=.8);
#plt.plot([1,0],[0,1], linestyle='--', lw=2, color='r',alpha=.8)
```

```python
# Find  AUC   #######
area=sklearn.metrics.auc(num, deno)
print("area =", area)

print('Number of actual signal events in test sample:', len(data_test[data_test.Labels == 's']))
print('ML Prediction SIG:', nsignal[nsignal/nbackground > 10])
print('Number of actual background events in test sample:', len(data_test[data_test.Labels == 'b']))
print('ML Prediction BKGD:', nbackground[nsignal/nbackground > 10])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Convert predictions to binary (threshold at 0.5)
binary_preds = (predictions > 0.5).astype(int)

# True labels
y_true = data_test.Labels.cat.codes

# Compute metrics
accuracy = accuracy_score(y_true, binary_preds)
precision = precision_score(y_true, binary_preds)
recall = recall_score(y_true, binary_preds)
f1 = f1_score(y_true, binary_preds)
auc = roc_auc_score(y_true, predictions)  # AUC uses raw probabilities

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUC: {auc:.4f}")
```

```python
from sklearn.calibration import calibration_curve

prob_true, prob_pred = calibration_curve(y_true, predictions, n_bins=10)

plt.plot(prob_pred, prob_true, marker="o", label="XGBoost")
plt.plot([0, 1], [0, 1], linestyle="--", label="Perfect Calibration")
plt.xlabel("Predicted Probability")
plt.ylabel("Actual Probability")
plt.title("Calibration Curve")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_true, binary_preds)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_true, predictions)

plt.plot(thresholds, precisions[:-1], label="Precision")
plt.plot(thresholds, recalls[:-1], label="Recall")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Precision-Recall vs Threshold")
plt.legend()
plt.show()
```

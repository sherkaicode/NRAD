```python
feature_names = data_train.columns[:-2].tolist()
feature_names = [str(f) for f in feature_names]# we skip the first and last two columns because they are the ID, weight, and label
train = xgb.DMatrix(data=data_train[feature_names],label=data_train.Labels.cat.codes,missing=-999.0,feature_names=feature_names)
test = xgb.DMatrix(data=data_test[feature_names],label=data_test.Labels.cat.codes, missing=-999.0,feature_names=feature_names)
```

```python
import numpy as np

def asimov_significance(y_true, y_pred, threshold=0.5):
    """
    Calculate Asimov significance given true labels and predicted scores.

    Parameters:
    y_true : np.array
        True labels (1 for signal, 0 for background)
    y_pred : np.array
        Predicted probabilities or scores from the BDT
    threshold : float
        Decision threshold to classify signal vs background

    Returns:
    float
        Asimov significance Z
    """
    # Convert probabilities into binary predictions
    y_classified = (y_pred >= threshold).astype(int)
    
    # Count signal and background events
    S = np.sum((y_true == 1) & (y_classified == 1))  # True Positives (Signal)
    B = np.sum((y_true == 0) & (y_classified == 1))  # False Positives (Background)
    # print(S)
    # print(B)
    
    if B == 0:  # Avoid division by zero
        return 0
    
    # Compute Asimov significance
    # Z = np.sqrt(2 * ((S + B) * np.log(1 + S / B) - S))
    Z = S/np.sqrt(B)
    
    return Z

# Example usage
thresholds = np.linspace(0, 1, 100)
y_true = test.get_label()
y_pred = booster.predict(test)

Z = asimov_significance(y_true, y_pred, threshold=0.5)
Z_vals = []
# Initial
# print(S_initial, B_initial)
S_initial = np.sum(test.get_label())
B_initial = len(test.get_label()) - S_initial
plt.axhline(y = S_initial/np.sqrt(B_initial), label = "Initial")
for threshold in thresholds:
    Z_vals.append(asimov_significance(y_true, y_pred, threshold=threshold))

plt.scatter(thresholds, Z_vals)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

# Get predictions for training and test data
train_preds = booster.predict(train)
test_preds = booster.predict(test)

# Split predictions by true class
train_signal = train_preds[train.get_label() == 1]
train_background = train_preds[train.get_label() == 0]
test_signal = test_preds[test.get_label() == 1]
test_background = test_preds[test.get_label() == 0]

# Perform KS test for signal and background distributions
ks_signal, p_signal = ks_2samp(train_signal, test_signal)
ks_background, p_background = ks_2samp(train_background, test_background)

# Plot distributions of predictions
plt.figure(figsize=(8, 6))

plt.hist(train_signal, bins=50, alpha=0.5, color='blue', label='Train Signal', density=True)
plt.hist(test_signal, bins=50, alpha=0.5, color='red', label='Test Signal', density=True, linestyle="--")

plt.hist(train_background, bins=50, alpha=0.5, color='green', label='Train Background', density=True)
plt.hist(test_background, bins=50, alpha=0.5, color='orange', label='Test Background', density=True, linestyle="--")

plt.xlabel("BDT Score")
plt.ylabel("Density")
plt.title("Overtraining Check: Training vs. Test Score Distributions")
plt.legend()
plt.grid()
plt.show()

# Print KS test results
print(f"KS Statistic (Signal): {ks_signal:.4f}, p-value: {p_signal:.4f}")
print(f"KS Statistic (Background): {ks_background:.4f}, p-value: {p_background:.4f}")

# Interpretation
if ks_signal > 0.1 or ks_background > 0.1:  # Threshold can be adjusted
    print("\n⚠️ Warning: Overfitting detected! The BDT score distributions differ significantly between training and test sets.")
else:
    print("\n✅ No significant overfitting detected. The training and test score distributions are similar.")
```
